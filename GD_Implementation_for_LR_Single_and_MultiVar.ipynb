## LR Full Implementation

### Step1: Initialize parameters (theta_0 & theta_1) with random value or simply zero. Also choose the Learning rate.

```python
th_0=0
th_1=0
learning_rate= 0.0001
```

### Step2: Use (theta_0 & theta_1) to predict the output h(x)= theta_0 + theta_1 * x.
#### Note: you will need to iterate through all data points

```python
H_x=th_0 + th_1 *X
print("h(x):", H_x)
```

### Step3: Calculate the MSE Cost function 𝑱(theta_0,theta_1 ).
<b>Note:</b> You better use either dot product or norm square of the error vector.

```python
m= len(X)
error = H_x - y
print(error)
mse= (1/(2*m)) * error@error
print("j=",mse)
```

### Step4: Calculate the gradient components for theta_0 and theta_1.
<b>Note:</b> You can use the error vector calculated in the previous step.

```python
gred_th_0 = 1/m * np.sum(error)
gred_th_1= 1/m * np.sum(error*X)
print("deriv_th_0 = ",gred_th_0)
print("deriv_th_1 = ",gred_th_1)
```

### Step5: Update the parameters (simultaneously).

```python
update_th_0 = th_0 - learning_rate* gred_th_0
updated_th_1 = th_1 - learning_rate * gred_th_1
print("theta_0_new =", update_th_0)
print("theta_1_new =", updated_th_1)
```

### Step6: Repeat from 2 to 5 until converge to the minimum or achieve maximum iterations.
#### The objective from this step is to combine all the previous steps and iterate untill you either achieve the maximum number of iterations or reach the minimum.

```python
max_iter = 1000
m= len(X)
th_0=0
th_1 = 0
learning_rate= 0.01

for itr in range(max_iter):
  y_pred = th_0 +th_1 *X
  print(y_pred)
  print("    ")

  error = y_pred - y
  print(error)
  print("    ")
  
  mse= (1/(2*m)) * error@error
  print("j=",mse)
  print("    ")

  gred_th_0=(1/m)*np.sum(error)
  gred_th_1=(1/m) * np.sum(error*X)
  print("d_theta_0 :",gred_th_0)
  print("d_theta_1 :",gred_th_1)
  print("    ")
  
  th_0 = th_0 - learning_rate * gred_th_0
  th_1 = th_1 - learning_rate * gred_th_1
  print("theta_0_new :",th_0)
  print("theta_1_new :",th_1)
  print("*********************************************************")



```

```python
max_iter = 1000
m = len(X)
th_0 = 0
th_1 = 0
learning_rate = 0.01
tolerance = 0.1

for itr in range(max_iter):
    y_pred = th_0 + th_1 * X
    print("Predictions:", y_pred)

    error = y_pred - y
    print("Error:", error)

    mse = (1 / (2 * m)) * np.dot(error, error)
    print("MSE:", mse)

    gred_th_0 = (1 / m) * np.sum(error)
    gred_th_1 = (1 / m) * np.sum(error * X)

    print("Gradient for theta_0:", gred_th_0)
    print("Gradient for theta_1:", gred_th_1)

    gradient_vector = np.array([gred_th_0, gred_th_1])

    gradient_norm = np.linalg.norm(gradient_vector)
    print("Gradient Norm:", gradient_norm)
    if gradient_norm < tolerance:
        print("Convergence reached.")
        break

    th_0 = th_0 - learning_rate * gred_th_0
    th_1 = th_1 - learning_rate * gred_th_1

    print("Updated theta_0:", th_0)
    print("Updated theta_1:", th_1)

    print("*********************************************************")
```

#### Predict y values using the LR equation
##### h(x)= theta_0 + theta_1 * x

```python
H_x = th_0 + th_1 * X
print("h(x) = y_predict:", H_x)
print("y_actual:", y)
```

## Plot loss function
### Repeat your implementaion version and save loss for each iteration (epoch)

```python
th_0= 0
th_1 = 0 
learning_rate = 0.01
max_iter=1000
loss_hist= []
for itr in range(max_iter):
    print(f"************ Iteration {itr} *************")
    print("    ")
    h_x = th_0 + th_1 * X
    print(f"h(x): {h_x}")
    print("    ")
    e = h_x - y
    print(f'Error Vector: {e}')
    print("    ")
    m = len(y)
    j = (np.linalg.norm(e) ** 2) / (2*m)
    loss_hist.append(j)

    print(f"j: {j}")
    print("       ")
    gradient_vector = (1/m) * e @ np.array([np.ones(X.shape),X]).T
    print(f"Gradient Vector: {gradient_vector}")
    print("    ")
    print(f"Gradient Vector Norm: {np.linalg.norm(gradient_vector)}")

    print("    ")
    th_0 = th_0 - learning_rate * gradient_vector[0]
    th_1 = th_1 - learning_rate * gradient_vector[1]
    print(f"Theta_0_new: {th_0}")
    print(f"Theta_1_new: {th_1}")
    if (np.linalg.norm(gradient_vector)) < 0.1:
        break
    print("       ")
```

### Generalize your implementation to work for MultiVariables
#### Vectorize your implementation.
<b>Notes:</b>
- You need to add column that contains ones to the data. This column represent x feature for theta_0.
- Use learninig rate = 0.0001.

```python
X2 = data2[:,:3]
y2= data2[:,3]
X = np.c_[np.ones(X2.shape[0]),X2]
max_itr= 1000
m = len(y2)
theta = np.zeros(X2.shape[1])
learning_rate = 0.0001
tolerance = 0.1
loss_history=[]
for itr in range(max_iter):
    y_pred = X2.dot(theta)
    error = y_pred - y2
    MSE = (1/2*m)* np.dot(error , error)
    loss_history.append(MSE)
    gradient = (1 / m) * X2.T.dot(error)

    gradient_norm = np.linalg.norm(gradient)

    print(f"****************** Iteration {itr} ********************")
    print("h(x):", y_pred)
    print("Error Vector:", error)
    print("j =", mse)
    print("Gradient Vector:\n", gradient)
    print("Gradient Vector Norm:", gradient_norm)
    theta = theta - learning_rate * gradient

    print(f"theta_new : {theta}")
    print("*********************************************************")

    if gradient_norm < tolerance:
        print("Convergence reached.")
        break
print("\n****************** Training Report ********************")
print(f"Gradient Descent converged after {itr} iterations\n")
print(f"Optimal theta values : {theta}")
print("\nError Vector:\n", error)
print("Final Cost =", mse)
print("\nh(x) = y_predict:\n", y_pred)
print("y_actual:\n", y)
```

#### Predict y values using the LR equation
##### h(x)= theta_0 + theta_1 * x1 + theta_2 * x2 + theta_3 * x3

```python
y_pred = X2.dot(theta)
print("\nPredicted y values (h(x)):")
print(y_pred)
print("\nActual y values:")
print(y2)
```

### Plot loss vs. iterations

```python
plt.plot(range(len(loss_history)), loss_history, label="cost (loss)")
plt.xlabel("Iterations")
plt.ylabel("Loss (MSE)")
plt.title("Loss vs. Iterations")
plt.show()
```
